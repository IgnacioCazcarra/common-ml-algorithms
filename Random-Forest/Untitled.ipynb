```python
from sklearn.datasets import make_classification
import matplotlib.pyplot as plt
```

### Generate mock data

```python
x,y = make_classification(n_samples=1000, n_classes=3, n_informative=4)
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.3, random_state=1)
```

### Build and train the model

```python
from sklearn.ensemble import RandomForestClassifier

clf = RandomForestClassifier(max_depth = 50, n_estimators = 150, max_leaf_nodes = 100)
```

```python
clf.fit(x_train, y_train)

preds = clf.predict(x_test)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test,preds))
```

### Visualization example

#### (For the sake of the example we'll build a much less complex classifier)

```python
estimators = 5
depth = 5

clf_example = RandomForestClassifier(max_depth = depth, n_estimators = estimators, max_leaf_nodes = 5)
```

```python
clf_example.fit(x_train, y_train)
```

```python
from sklearn import tree
```

```python
fig, axes = plt.subplots(nrows = 1,ncols = 5, figsize = (30,6))

for index in range(estimators):
    tree.plot_tree(clf_example.estimators_[index],
                   filled = True,
                   ax = axes[index]);

    axes[index].set_title('Tree: ' + str(index), fontsize = 40)
plt.show()
```

## As we can see, random forest is a collection of decision trees. Each tree root is decided by selecting 2 random features and comparing its gini. Then, the one with the smaller number becomes the root and we do the same thing for each tree
