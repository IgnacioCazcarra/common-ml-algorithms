```python
import pandas as pd
import numpy as np
from sklearn.datasets import make_regression
import matplotlib.pyplot as plt
```

## Generate mock data and visualize it

```python
x,y = make_regression(n_samples=500, n_features=1, noise=10, random_state=1, effective_rank=1)

constant = 50

# Up the values so we dont get negative x's
x = [i+constant for i in x]
y = [i+constant for i in y]
```

```python
plt.scatter(x,y)
plt.show()
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.20, random_state=1)
```

### 1) Linear regression

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()

lr.fit(x_train,y_train)

preds = lr.predict(x_test)
```

```python
plt.scatter(x_test, y_test,color='b')
plt.plot(x_test, preds,color='r')
```
