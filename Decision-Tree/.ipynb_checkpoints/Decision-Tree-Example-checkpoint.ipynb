```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn
from sklearn.datasets import make_classification
```

### Generate mock data

```python
x,y = make_classification(n_samples=1000, n_classes=3, n_informative=4)
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.3, random_state=1)
```

### Build and train the model

```python
from sklearn import tree 

clf = tree.DecisionTreeClassifier(max_leaf_nodes=50, max_features=4, max_depth=50)
```

```python
clf.fit(x_train,y_train)
```

```python
preds = clf.predict(x_test)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test,preds))
```

### Visualization of the tree

```python
fig = plt.figure(figsize=(100,100))

tree.plot_tree(clf, filled=True, fontsize=20) 

plt.show()
```

### Visualization example of small tree

```python
clf_example = tree.DecisionTreeClassifier(max_leaf_nodes=5, max_features=4, max_depth=5)
clf_example.fit(x_train, y_train)
```

```python
fig = plt.figure(figsize=(20,20))

tree.plot_tree(clf_example, filled=True, fontsize=20) 

plt.show()
```
