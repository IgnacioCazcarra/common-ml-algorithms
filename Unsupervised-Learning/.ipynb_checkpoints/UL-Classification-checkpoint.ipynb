```python
from sklearn.datasets import make_classification
```

### Generate mock data

#### Since we are working with unsupervised algorithms, we wont use 'y'

```python
x,y = make_classification(n_samples=1000)

x.shape
```

### We transform the data into 2d with PCA

```python
from sklearn.decomposition import PCA

pca = PCA(2)

x = pca.fit_transform(x)
```

### 1) K-Means

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters = 5)
```

```python
preds = kmeans.fit_predict(x)
```

```python
import numpy as np

unique_classes = np.unique(preds)

centroids = kmeans.cluster_centers_
```

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(10,10))

for i in unique_classes:
    plt.scatter(x[preds==i,0] , x[preds==i,1] , label=i, alpha=0.6)

plt.scatter(centroids[:,0], centroids[:,1], s=100, color="black")    
    
plt.show()
```
